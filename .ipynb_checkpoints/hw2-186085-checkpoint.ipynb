````python
import sys
from des import DesKey
````

````python
def test_des(key_text, message, initial):
    key = DesKey(key_text)
    print("======================")
    print(f"DES: {key.is_single()}") # is it single DES or triple 3DES
    encrypted = key.encrypt(message, initial=initial) # check if length is divisible by 8
    # decrypted = key.decrypt(encrypted, initial=initial)

    print(f"> the key is")
    print(key_text)
    print('\n'.join(format(byte, '08b') for byte in key_text))
    print(f"and the size in bytes is {len(key_text)}")
    print("> the message is")
    print(message)
    print('\n'.join(format(byte, '08b') for byte in message))
    print(f"and the size in bytes is {len(message)}")
    print("> the encrypted message is:")
    print(encrypted)
    print('\n'.join(format(byte, '08b') for byte in encrypted))
    # print("> the decrypted message is:")
    # print(decrypted)
    print("======================")
````

## Пример 1 - клуч 0 порака 0

````python
####### test output when key and message are 0s.
keyA = b"\x00"*8
initialA = b"\x00"*8
messageA = b"\x00"*32
test_des(keyA, messageA, initialA)
````

## Пример 2 - клуч 1 порака 1

````python
####### test round keys of a key of all 1s.
keyB = b"\xFF"*8
initialB = b"\xFF"*8
messageB = b"\xFF"*32
test_des(keyB, messageB, initialB)
````

## Пример 3 - клуч 0 порака 0 со промена во 1 

### Промена на бит во првите 8 бајти од пораката

````python
####### test output when key is 0 and message has 1 bit changed in the first 8 bytes
keyC = b"\x00"*8
initialC = b"\x00"*8
messageC = b"\x00"*32
modMessageC = bytearray(messageC)
modMessageC[0] ^= 0b01000000
test_des(keyC, bytes(modMessageC), initialC)
````

Кога проментата во пораката се случува во првите 8 бајти, тогаш се открива првиот дел од пораката. Аваланш ефектот делува на скоро половина од битовите. Резултатот е скоро идентичен со тој од во првиот пример.

### Промена на бит во вторите 8 бајти од пораката

````python
####### test output when key is 0 and message has 1 bit changed in the second 8 bytes
keyC = b"\x00"*8
initialC = b"\x00"*8
messageC = b"\x00"*32
modMessageC = bytearray(messageC)
modMessageC[8] ^= 0b01000000     # changed to modMessageC[8]
test_des(keyC, bytes(modMessageC), initialC)
````

Сега промената е направена во бит кој се наоѓа во вторите 8 бајти од пораката. Ефектот на аваланш е поголем.
